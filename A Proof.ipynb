Imagine we have $a$ number of bins. 

One bin for numbers divisible by $a$, $\leftarrow$ bin $A_0$

another for numbers $1$ more than numbers divisible by $a$, $\leftarrow$ bin $A_1$

another for numbers $2$ more than numbers divisible by $a$,$\leftarrow$ bin $A_2$

...........................................................,

............................................................,

another for numbers $a-1$ more than numbers divisible by $a$.$\leftarrow$ bin $A_{a-1}$

##### Step 0

>from $n\ge0$ we are able to fill $A_0$ from the expression $ax+by$, by $y=0$ and $x=0,1,\ldots$

Since $a$ and $b$ don't have common divisors except $1$, $b$ cannot be in bin $A_0$.

##### Step 1
>So from $n\ge b$ fill $A_{b \bmod a}$ from the expression $ax+by$, by $y=1$ and $x=0,1,\ldots$

Now we have considered 2 bins.

Then we can think where would $2b$ go. It can't go to $A_0$ because before $ab$ we can't find a multiply of $b$ that can be divided by $a$ (We are only considering $b,2b,3b,\ldots$). $2b$ can't go to bin $A_{b\bmod a}$ either. 

let $b=aq_1+r_1$ [1] where $q_1$ and $r_1$ are the quotient and remainder ($b\bmod a$). 

If $2b$ goes to $A_{b\bmod a}$, then we can say $2b=aq^{/}+r_1$ [2]

Then by [2]-[1] we get, $b=a(q^/-q_1)$ which is a contradiction. So $2b$ should be in some other bin than $A_0$ and $A_{r_1}$

Let's say that bin is $A_{r_2}$ where $r_2=2b\bmod a$

##### Step 2
>So from $n\ge 2b$ we can fill $A_{r_2}$ from the expression $ax+b$, by $y=2$ and $x=0,1,\ldots$



Let's now think of $3b$. Should it go to a new bin or an already filled one? By following the same argument as for $3b$ we can say, $3b$ can't go to $A_0$ or $A_{r_2}$. Can it go to bin $A_{r_1}$? If $3b$ goes to $A_{r_1}$ we can show that $2b$ is divisible by $a$ which is a contradiction. So $3b$ goes to a whole new bin. Let's call it $A_{r_3}$

##### Step 3
> So from $n\ge3b$ we can fill $A_{r_3}$ from $ax+by$, by $y=3$ and $x=0,1,\ldots$

##### Step (a-1)
Likewise we can do up to $(a-1)b$. In that step we fill bin $A_{r_{a-1}}$ which is also a whole new bin which was not filled before.

So After that Let's examine our $a$ number of bins.

$A_0$ contains all numbers of the form $ax$

$A_{r_1}$ contains all numbers of the form $ax+b$

.

.

.

$A_{r_{a-1}}$ contains all numbers of the form $ax+(a-1)b$

By the time we start filling $A_{r_{a-1}}$ the other bins are already filled, there will be no empty bin.

Imagine we strike through the number in the number line when we fill that to one of the bins. 

So from step 0, $\{0,a,2a,3a,\ldots\}$ will be gone from the number line.

From step 1, $\{b,a+b,2a+b, \ldots\}$ will be gone.

.

.

.

From step (a-1), $\{(a-1)b, a+(a-1)b, 2a+(a-1)b,\ldots\}$ will be gone.

By the time we fill $(a-1)b$, there will be no $(a-1)b-1$ on the number line because it will be greater than any of $b, 2b,3b,\ldots, (a-2)b$ and it is not in the bin $A_{r_{a-1}}$ (it must have remainder other than $r_{a-1}$)

But what about $(a-1)b-a$? Is it in the number line or not. It should be in the number line because it's less than $(a-1)b$ but have a remainder $r_{a-1}$. But the numbers between $(a-1)b-a$ and $(a-1)b$ are filled.

That means from $(a-1)b-a+1$ onwards, we can generate any number from $ax+by$, $x,y\ge0$

If you look carefully you'll see that $(a-1)b-a+1$ is that `magical` number $(a-1)(b-1)$.
